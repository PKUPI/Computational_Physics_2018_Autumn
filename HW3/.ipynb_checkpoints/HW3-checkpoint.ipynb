```python
import numpy as np
import matplotlib.pyplot as plt
import os
import math
import time
from scipy.stats import chi2
```

```python
'''
Householder QR分解
'''

def buildP(M,i):
    '''
    这个函数返回一个函数句柄P，P(row)的返回值为：将M构造的P_(i)左乘row得到的矢量
    '''
    A = M
    dim = A.shape[0]
    w = np.array(A[i,i:])
    l = np.linalg.norm(w)
    w[0] -= l
    w_norm = np.linalg.norm(w)**2
    
    def P(x):
        y = np.array(x)
        y[i:] = x[i:] - (2/w_norm) * np.dot(w,x[i:])* w
        return y
    return P

def HouseholderQR(M):
    '''
    Householder QR 分解
    input: M, np.array
    output: Q, np.array
            R,np.array
    '''
    A = M
    dim = np.shape(A)[0]
    Q = np.eye(dim)
    for i in range(dim-1):
        # 构造一个P矩阵，但由于计算的特点，不需要真的构造完整的矩阵并进行矩阵乘
        P = buildP(A,i)
        # 在处理第i列的时候，0到i-1行都不发生变化，i到n行发生变化
        # 而之前的列在i到n行都是0，故也不发生变化
        # 所以只处理i:n行和i:n列
        A[i:dim] = [P(A[j]) for j in range(i, dim)]
        Q[:] = [P(Q[j]) for j in range(dim)]
    return Q.T,A.T

dim = input("Input the dimension of matrix:")
M = np.random.rand(int(dim),int(dim))
Q,R = HouseholderQR(M)
print(Q)
print(R)
```

```python
def Givens(A, i, j):
    '''
    得到Givens转动的函数G，首先解出cos,sin，分别命名为c,s
    方程为 A[i,j] cosa + A[j,j] sin a = 0
          A[i,j] -sina + A[j,j] cos a = norm([A[i,j],A[j,j]])
    输入A:np.array
        i,j:需要旋转的两个方向的下标
    '''
    s,c = np.linalg.solve([[-A[j,i], A[j,j]],[A[j,j], A[j,i]]],[np.sqrt(A[j][j]**2 + A[j][i]**2 ),0])
    
    def G(x):
        a = x[j]*c-x[i]*s
        b = x[j]*s+x[i]*c
        # 交换
        x[j] = a
        x[i] = b
    return G


def GivensQR(B):
    '''
    Givens QR 分解
    input: B, np.array
    output: Q, np.array
            R,np.array
    '''
    A = np.array(B)
    dim = A.shape[0]
    Q = np.identity(dim)
   
    for j in range(dim-1):
        for i in range(j+1, dim):
            # 此处返回一个G函数，相当于对i,j方向进行旋转
            # 由于调用引用，G函数的操作会改变矩阵原始值
            G = Givens(A, i, j)
            for k in range(j-1, dim-1):
                G(A[k])
            G(Q)
    return Q.T, A.T

dim = input("Input the dimension of matrix:")
M = np.random.rand(int(dim),int(dim))
Q,R = GivensQR(M)
print(Q)
print(R)
```

```python
'''
测试上面两种QR分解的速度
根据电脑性能不同这块代码执行的时间会不一样。大约在几分钟左右。
'''

testcase = 64
TestMatrix = np.random.rand(testcase,testcase,testcase)*2 -1
HouseholderTime = []
for dim in range(2,testcase):
    start = time.time()
    for i in range(testcase):
        HouseholderQR(TestMatrix[i,:dim,:dim])
    end = time.time()
    HouseholderTime.append(end-start)

GivensTime = []
for dim in range(2,testcase):
    start = time.time()
    for i in range(testcase):
        GivensQR(TestMatrix[i,:dim,:dim])
    end = time.time()
    GivensTime.append(end-start)

state_of_art = []
for dim in range(2,testcase):
    start = time.time()
    for i in range(testcase):
        np.linalg.qr(TestMatrix[i,:dim,:dim])
    end = time.time()
    state_of_art.append(end-start)
```

```python
plt.figure(figsize=(16,10))

plt.plot(range(len(HouseholderTime)),HouseholderTime,marker='o',label='Householder')
plt.plot(range(len(GivensTime)),GivensTime,marker='o',label='Givens')
plt.plot(range(len(state_of_art)),state_of_art,marker='o',label='NumPy Implementation')
plt.grid()
plt.legend()
plt.savefig('./compare.png')
plt.show()
```

```python
'''
幂次法求本征值
'''
def solveEigen(dim):
    A = np.zeros((dim,dim))
    for i in range(dim):
        for j in range(dim):
            A[i,j] = - ((dim+i-1) % dim ==j) - ((i+1)%dim == j) + 2*(i==j)
    q = np.random.rand(dim)
    q /= np.linalg.norm(q)
    _q = np.zeros_like(q)
    l = []
    while np.max(np.abs(q-_q))>1e-10:
        _q = q 
        q = np.dot(A,q)
        q /= np.linalg.norm(q)
        l.append(np.dot(q.T,A).dot(q))
    
    plt.figure()
    plt.plot(range(len(l)),l,'o-')
    plt.grid()
    plt.savefig('./power.png')
    plt.show()
    
    return q,l[-1]

ans,lam = solveEigen(10)
print(ans)
print(lam)
```

```python
'''
读取数据
数据应当存放在./data文件夹中
'''
def loadData():
    '''
    读取数据
    return: traj, np.array N * D
            N: block数目，也就是组态数目
            D: 时间片数目
    '''
    traj = np.zeros((200,32))
    N_cnt = 0
    for i in range(100,320):    
        filename = "./data/traj_"+str(i)+"0_pion.txt"
        with open(filename,"r") as f:
            if os.path.getsize(filename)==0:
                continue
            j = 0
            for line in f.readlines():
                traj[N_cnt,j]=float(line.split(' ')[2])
                j+=1
            N_cnt += 1
    return traj

traj = loadData()
```

```python
'''
3.(a)
C_mean: (D,) np.ndarray,对每个组态求平均(axis=0)
C_delta: (D, ) np.ndarray,对每个组态求误差
C_error: (D, ) np.ndarray，相对误差(百分误差)
'''
C_mean = traj.mean(axis = 0)

C_delta = np.sqrt(1. /(200 *199) *
        np.sum(np.square(traj - C_mean.reshape((1,-1))),axis = 0))
C_error = C_delta / C_mean *100


plt.figure()
plt.plot(range(32),C_error,'.-')
plt.grid()
plt.xlabel('t')
plt.ylabel('Relative Error (%)')
plt.savefig('./relativeError.png')
plt.show()
```

```python
'''
3.(b)
m_eff: 按照式子(9)计算得到的有效质量
t0: 误差超过信号的时间片（人为取定）
m_eff_jk : N * D 利用Jackknife方法进行计算，N: jk方法得到的数据的组数，D: 数据的时间片数
traj_jk : 剔除某一个组态后的数据(N-1)*D, N:原来组态数, D:原来的数据时间片数
C_mean_jk :用上面的traj_jk计算得到的C平均值，(D,) D:时间片数
m_eff_delta: jk方法计算得到的有效质量的涨落
'''
m_eff = np.log(C_mean[:31] / C_mean[1:]) / np.log(np.e)
t0 = 25
m_eff_jk = np.zeros((200,31))

for i in range(200):
    traj_jk = np.zeros((199,32))
    traj_jk[:i,:] = traj[:i,:]
    traj_jk[i:,:] = traj[i+1:,:]
    C_mean_jk = traj_jk.mean(axis = 0)
    m_eff_jk[i,:] = np.log(C_mean_jk[:31] / C_mean_jk[1:]) / np.log(np.e)

m_eff_delta = np.sqrt((200-1)/200. *np.sum(np.square(m_eff_jk - m_eff.reshape(1,-1)),axis = 0))


# 下面是画图
plt.figure()
plt.errorbar(range(t0),m_eff[:t0],yerr=m_eff_delta[:t0],ecolor='black',fmt='.-')
plt.xlabel('t')
plt.ylabel('Effective mass')
plt.grid()
plt.savefig('./effm1.png')
plt.ylim(0.23,0.25)
plt.plot(range(t0),np.ones(t0)*0.24475327206230157,'g-.')
plt.plot(range(t0),np.ones(t0)*0.24153859460449217,'g-.')
plt.plot(range(t0),np.ones(t0)*0.24316705723146828,'g-')
plt.plot(range(12, 16),np.ones(4)*0.24475327206230157,'r-')
plt.plot(range(12, 16),np.ones(4)*0.24153859460449217,'r-')
plt.plot(range(12, 16),np.ones(4)*0.24316705723146828,'r-')
plt.savefig('./merror1.png')
plt.show()
```

```python
'''
3.(c)
'''
def solve(f,left,right,verbose=False, e =1e-8):
    '''
    二分法解方程
    '''
    if verbose:
        print(f(left),f(right))
    if f(left)* f(right)>0:
        raise Exception("Invalid range!")
        return
    mid = (left+right)/2.
    if abs(right-left)<e:
        return mid
    if f(left)*f(mid)<0:
        return solve(f,left,mid,verbose=verbose)
    if f(mid)*f(right)<0:
        return solve(f,mid,right,verbose=verbose)

    
m_fit = np.zeros((32,32))
kai2 = np.zeros((32,32))
kai2_per_dof = np.ones((32,32)) * 0xffff

for t_min in range(5,25):
    for t_max in range(t_min+3,25):
        dof = t_max - t_min
        # 解出让Kai^2最小的m
        m_fit[t_min,t_max] = np.sum((m_eff[t_min:t_max] / m_eff_delta[t_min:t_max]))/np.sum(1/m_eff_delta[t_min:t_max])
        # 求解p-value
        kai2[t_min,t_max] = np.sum(np.square((m_eff[t_min:t_max]-m_fit[t_min,t_max])/m_eff_delta[t_min:t_max]))
        kai2_per_dof[t_min,t_max] = kai2[t_min,t_max]/(t_max-t_min)


def find_min(M):
    m,n = np.shape(M)
    # 逐行找最小值
    min_idx = np.argmin(M,axis=1)
    # 比较这些最小值，找到每行最小值里面最小的的行号
    min_row = np.argmin(M[range(m),min_idx])
    # 最小值的列号
    min_col = min_idx[min_row]
    return min_row,min_col,M[min_row,min_col]

# 找到最小值和对应的行列号
row,col,minval = find_min(kai2_per_dof)
print("t_min,t_max, chi2/d.o.f:",find_min(kai2_per_dof))

# 计算Q
Q = chi2.cdf(kai2[row,col],col-row)
print("Best fit mass:",m_fit[row,col])
print("Chi2:",kai2[row,col])
print("Q,p-value",Q,1-Q)

# 计算1sigma对应的chi2值
f = lambda x: chi2.cdf(x,dof)-0.683
kai2_sig = solve(f,3.5,20)
#print(kai2_sig)

# 计算上述chi2值对应的m_eff区间
g = lambda m: np.sum(np.square((m_eff[row:col]-m)/m_eff_delta[row:col]))-kai2_sig
print("Upper bound:",solve(g,0.243151,0.245444))
print("Lower bound:",solve(g,0.241551,0.241451))
print(solve(g,0.243151,0.245444)-solve(g,0.241551,0.241451))
```
