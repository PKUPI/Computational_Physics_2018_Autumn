```python
'''
Runge Effect.
'''
x = np.array(range(-10,11,1))/10.0
_x = np.array(range(-20,21,1))/20.0
f = lambda x: 1./(1+25 * x**2) 
```

```python
'''
牛顿法进行多项式插值
'''
def solveL(L,b):
    '''
    解下三角线性方程组，要求L为下三角方阵，Lx=b，返回x
    return type: np.array()
    ''' 
    L = np.array(L); b = np.array(b);
    n = np.shape(L)[0]
    x = copy.deepcopy(b)
    for i in range(n-1):
        x[i] /= L[i][i]
        x[i+1:n] = x[i+1:n] - x[i]*L[i+1:n,i]
    x[n-1] /= L[n-1,n-1]
    return x

def solveU(U,b):
    '''
    回代法求解上三角线性方程组，要求U为上三角方阵，Ux=b，返回x
    return type: np.array
    '''
    U = np.array(U); b = np.array(b)
    n = np.shape(U)[0]
    x = copy.deepcopy(b)
    i = n-1
    while i >= 0:
        x[i] /= U[i][i]
        for j in range(i):
            x[j] -= U[j][i]*x[i]
        i -= 1 
    x[0] /= U[0][0]
    return x


def constructCoe(x):
    '''
    要求a，需要解方程Ca = y
    C是由采样点序列x定义的系数矩阵，具体定义见刘川讲义P24式子(3.7)；
    y是由采样点序列x计算得到的函数值序列y；
    这个函数实现的功能是：输入采样点序列x，构造求系数矩阵
    input type: np.array,shape:(dim,)
    return type: np.array,shape==(dim,dim)
    '''
    dim = np.shape(x)[0]
    C = np.zeros((dim,dim))
    C[:,0] = np.ones_like(C[:,0])
    for i in range(1,dim):
        C[:,i] = (x - x[i-1])* C[:,i-1]
        C[0:i-1,i] = 0
    return C

def Newton(sample_x,sample_y,x):
    '''
    构造牛顿法插值的结果
    input:
        sample_x: 采样点序列x, np.array, shape:(dim,)
        sample_x: 采样点序列x的取值, np.array, shape:(dim,)
        x:待求值的点序列x, np.array
    return:
        poly: np.polynomial.Polynomials
    '''
    dim = np.shape(sample_x)[0]
    a = solveL(constructCoe(sample_x),sample_y)
    result = np.zeros_like(x)
    n = 1
    for i in range(dim):
        result += a[i] * n
        n *= (x-sample_x[i])
    return result
```

```python
'''
切比雪夫多项式插值
'''
dim = 20
_x = np.cos((np.pi*(np.array(range(40))/2.+1./2)/20.))

def construct_c(dim):
    '''
    这个函数的目的是为了构造切比雪夫多项式的系数c_i，利用刘川讲义P33 式子3.43
    输入:dim: scalar，表示采样点数
    输出：c: np.array, shape:(dim,)
    '''
    sample_x = np.cos((np.pi*(np.array(range(dim))+0.5)/dim))
    c = np.zeros((dim,dim))
    for i in range(dim):
        for j in range(dim):
            c[i,j] = np.cos((i * np.pi * (j+ 0.5 ) / dim))
    c = 2 * np.dot(c,f(sample_x)) / dim
    return c

def Chebyshev(dim,x):
    '''
    这个函数的目的是为了计算切比雪夫多项式的值，利用Clenshaw算法进行展开，根据刘川讲义P34 式子3.50
    输入:dim: scalar，表示采样点数
        x: np.array，需要求切比雪夫多项式值的x值组成的向量
    输出：result: np.array, shape:x.shape
    '''
    c = construct_c(dim)#构造系数
    b = np.zeros((dim+2,np.shape(x)[0]))
    for i in range(dim-1,0,-1):
        b[i,:] = c[i] + 2 * x * b[i+1,:] - b[i+2,:]
    return 0.5 * c[0] +  x * b[1,:] - b[2,:]
```

```python
'''
三次样条插值并且实现
'''
# 定义插值的参数和采样点
num_sample = 21
interval = 2./(num_sample-1)
x = np.array(range(-10,11,1))/10.0
_x = np.array(range(-20,21,1))/20.0
f = lambda x: 1./(1+25 * x**2)
df = lambda x: -50.*x/((1+25 * x**2)**2)
y = f(x)

# 构造三对角矩阵元，生成a1-an,b2-bn,c1-c_{n-1}，并计算alpha,beta用于生成L矩阵和U矩阵
# 在这个问题中，每个采样点间隔是相等的，
# 所以只需要用interval一个参数，而不需要维护各个点之间间隔的数组
beta = interval * np.ones(num_sample)/6.
alpha = 2 * interval * np.ones(num_sample)/3.
c = interval * np.ones(num_sample)/6.
alpha[0] = - interval / 3.
c[0] = - interval / 6.
alpha[num_sample-1] = interval / 3.
beta[num_sample-1] = interval / 6.

b = np.zeros(num_sample)
b[0] = df(x[0]) - (y[1] - y[0])/interval
b[num_sample-1] = df(x[num_sample-1]) - (y[num_sample-1] - y[num_sample-2])/interval
for i in range(1,num_sample-1):
    b[i] = (y[i+1] + y[i-1] - 2 * y[i]) / (interval)

for i in range(1,num_sample):
    beta[i] = beta[i] / alpha[i-1]
    alpha[i] = alpha[i] - beta[i] * c[i-1]

def constructU(alpha,c):
    '''
    解三对角方程，构造上三角矩阵
    return U, np.array, shape:(alpha.shape[0],alpha.shape[0])
    '''
    dim = np.shape(alpha)[0]
    U = np.diag(alpha)
    for i in range(dim -1):
        U[i,i+1] = c[i]
    return U

def constructL(alpha,beta):
    '''
    解三对角方程，构造下三角矩阵
    return: L, np.array, shape:(alpha.shape[0],alpha.shape[0])
    '''
    dim = np.shape(alpha)[0]
    L = np.eye(dim)
    for i in range(dim-1):
        L[i+1,i] = beta[i+1]
    return L

# 解三对角矩阵得到三次样条插值的各个节点处的矩
M = solveU(constructU(alpha,c),solveL(constructL(alpha,beta),b))
print(M)
# interplot_x是三次样条插值得到的函数值，需要分段计算
interplot_x = np.zeros(2 * num_sample - 1)

# 通过计算出来的M的值，分段计算每个点的样条函数值
for i in range(np.shape(interplot_x)[0]):
    idx = int(i/2)
    if i%2==0:
        interplot_x[i] = y[idx]
    else:
        interplot_x[i] = M[idx] * (x[idx+1] - _x[i]) ** 3 / (6 * interval) + \
                         M[idx+1] * (_x[i] - x[idx]) ** 3 / (6 * interval) + \
                        ((y[idx+1]-y[idx])/interval - (interval / 6.) * (M[idx+1]-M[idx])) * (_x[i]-x[idx])+\
                        y[idx] - M[idx]* interval**2 / 6.
        
```

```python
def spine(num_sample,interval,x,y,dy,_x,k):
    '''
    由于本题要分别对X,Y进行三次样条插值，此处构造一个函数，
    此处代码直接从第一题中复制，流程完全一致
    arg:num_sample:采样点个数
        interval:采样点间隔（这里假设所有采样点是等间隔的）
        x：采样点
        y：采样点处的原函数值
        dy：采样点处原函数的一阶导数值（使用两端导数的边界条件）
        _x：待求值的点
        k：待求值的点的间隔是interval的1/k倍
    return: interplot_x: 插值后的函数值
            M： 插值函数在每个采样点的矩
    '''
    # 构造三对角矩阵元，生成a1-an,b2-bn,c1-c_{n-1}，并计算alpha,beta用于生成L矩阵和U矩阵
    # 在这个问题中，每个采样点间隔是相等的，
    # 所以只需要用interval一个参数，而不需要维护各个点之间间隔的数组
    beta = interval * np.ones(num_sample)/6.
    alpha = 2 * interval * np.ones(num_sample)/3.
    c = interval * np.ones(num_sample)/6.
    alpha[0] = - interval / 3.
    c[0] = - interval / 6.
    alpha[num_sample-1] = interval / 3.
    beta[num_sample-1] = interval / 6.

    b = np.zeros(num_sample)
    b[0] = dy[0] - (y[1] - y[0])/interval
    b[num_sample-1] = dy[num_sample-1] - (y[num_sample-1] - y[num_sample-2])/interval
    for i in range(1,num_sample-1):
        b[i] = (y[i+1] + y[i-1] - 2 * y[i]) / (interval)

    for i in range(1,num_sample):
        beta[i] = beta[i] / alpha[i-1]
        alpha[i] = alpha[i] - beta[i] * c[i-1]

    # 解三对角矩阵得到三次样条插值的各个节点处的矩
    M = solveU(constructU(alpha,c),solveL(constructL(alpha,beta),b))
    print(M)
    # interplot_x是三次样条插值得到的函数值，需要分段计算
    interplot_x = np.zeros(_x.shape[0])

    # 通过计算出来的M的值，分段计算每个点的样条函数值
    for i in range(_x.shape[0]):
        idx = int(i/k)
        if i%k==0:
            interplot_x[i] = y[idx]
        else:
            interplot_x[i] = M[idx] * (x[idx+1] - _x[i]) ** 3 / (6 * interval) + \
                             M[idx+1] * (_x[i] - x[idx]) ** 3 / (6 * interval) + \
                            ((y[idx+1]-y[idx])/interval - (interval / 6.) * (M[idx+1]-M[idx])) * (_x[i]-x[idx])+\
                            y[idx] - M[idx]* interval**2 / 6.

    return interplot_x,M

```

```python



def Int(f, a, b, n=1000):
    '''
    最后一项积分的被积函数求值
    '''
    h = (b-a)/(n*2) # 间隔
    return np.sum([h*( f(a+2*i*h) + 
                      4*f(a+(2*i+1)*h) + 
                      f(a+(2*i+2)*h))/3  
                     for i in range(n)])

def Int2(x):
    '''
    第二项常数项和瑕积分的计算
    '''
    ans = 0
    for i in range(1,18):
        ans += math.pow(x,i) / (math.factorial(i)*(i-0.5))
    ans *= np.pi * 0.5
  #  print("T2:",ans)
    return ans

def BinarySearch(f, a, b, e=1e-9):
    '''
    二分法求解方程，e=1e-9是一个默认的精确度，可以通过传参增大精确度
    '''
    if f(a)*f(b) > 0:
        return None
    if abs(a-b) < e:
        return a
    m = (a+b)/2
    if abs(f(m))<e:
        return m
    elif f(a)*f(m) < 0:
        return BinarySearch(f, a, m, e)
    elif f(m)*f(b) < 0: 
        return BinarySearch(f, m, b, e)


def f2(x):
    '''
    多项式部分，直接计算
    '''
    return np.pi**(3/2)*(1+0.25*x)

def g1(x, y, z, q):
    '''
    第一项求和，对各种可能的\overrightarrow{n}
    '''
    k = x*x+y*y+z*z
    return np.exp(-k)/(k-q)

def g2(t, q):
    '''
    具体计算的时候使用多了一项，可以提高精确度
    实际上只需要三项求和。
    '''
    return t**(-3/2)*np.exp(t*q)*(6*np.exp(-np.pi*np.pi/t)+
                                  12*np.exp(-2*np.pi*np.pi/t)+
                                  8*np.exp(-3*np.pi*np.pi/t)+
                                  6*np.exp(-4*np.pi*np.pi/t))

def Z00(x):
    T1 = 0
    for i in range(-4,5):
        for j in range(-4,5):
            for k in range(-4,5):
                T1 += g1(i, j, k, x)
    T1 *=np.exp(x)/(2*np.sqrt(np.pi))
    
    T2 = Int2(x)
    T3 = np.sqrt(np.pi) * Int(lambda t: g2(t, x), 0.2, 1)/2
    return T1+T2+T3-np.pi

print(BinarySearch(lambda x: Z00(x)-f2(x), 0.00001, 0.99999))
```
